```python
import matplotlib.pyplot as plt
import numpy as np

# Given values ( )
V_phase = 86.6  # V_phi,   (V)
Is = 4.7  #   (A)
Rs = 2.44  #    ()
Xs = 101.3  #   ()

# Calculate voltage drops (  )
V_Rs = Is * Rs  #    Rs (V)
V_Xs = Is * Xs  #    Xs (V)

# Phasor components for ohmic load (cos(phi)=1)
# For ohmic load, Is is in phase with V_phase.
# V_phase is along the real axis.
# V_phi = (V_phase, 0)
# Is*Rs is also along the real axis.
# j*Is*Xs is along the imaginary axis.

# Coordinates of the points for the vectors
# Origin ()
O = np.array([0, 0])

# Tip of V_phase vector (   V_phase)
P1 = np.array([V_phase, 0])

# Tip of V_phase + Is*Rs vector (   V_phase + Is*Rs)
P2 = np.array([V_phase + V_Rs, 0])

# Tip of E_A = V_phase + Is*Rs + j*Is*Xs vector (   E_A)
P3 = np.array([V_phase + V_Rs, V_Xs]) # This is E_A

# Create the plot ( )
plt.figure(figsize=(10, 8))

# Plotting vectors using quiver
# Vector V_phase ( V_phase)
plt.quiver(O[0], O[1], P1[0], P1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'$V_\\phi = {V_phase:.1f}V$')
# Vector Is*Rs ( Is*Rs)
plt.quiver(P1[0], P1[1], V_Rs, 0, angles='xy', scale_units='xy', scale=1, color='red', label=f'$I_sR_s = {V_Rs:.3f}V$')
# Vector j*Is*Xs ( j*Is*Xs)
plt.quiver(P2[0], P2[1], 0, V_Xs, angles='xy', scale_units='xy', scale=1, color='green', label=f'$jI_sX_s = j{V_Xs:.2f}V$')
# Vector E_A ( E_A)
plt.quiver(O[0], O[1], P3[0], P3[1], angles='xy', scale_units='xy', scale=1, color='purple', label=f'$E_A$')

# Annotate points ( )
plt.plot(O[0], O[1], 'ko') # Origin
plt.text(O[0] - 15, O[1] - 10, 'O', fontsize=12)

plt.plot(P1[0], P1[1], 'ko') # Tip of V_phase
plt.text(P1[0]/2, P1[1] + 10, '$V_\\phi$', color='blue', fontsize=12)

plt.plot(P2[0], P2[1], 'ko') # Tip of V_phase + Is*Rs
plt.text(P1[0] + V_Rs/2 -5 , P1[1] + 10, '$I_sR_s$', color='red', fontsize=12)

plt.plot(P3[0], P3[1], 'ko') # Tip of E_A
plt.text(P2[0] + 5, P2[1] + V_Xs/2, '$jI_sX_s$', color='green', fontsize=12)
plt.text(P3[0]/2 - 20, P3[1]/2 + 20, '$E_A$', color='purple', fontsize=14)


# Set axis limits for better visualization (     )
# Calculate magnitudes for E_A
E_A_mag = np.sqrt(P3[0]**2 + P3[1]**2)
plt.text(P3[0]*0.8, P3[1]*0.8 + 30, f'$|E_A| = {E_A_mag:.2f}V$', color='purple', fontsize=12)


max_x = max(P1[0], P2[0], P3[0]) + 50
max_y = max(P1[1], P2[1], P3[1]) + 50
min_x = min(0, P1[0], P2[0], P3[0]) - 50 # Ensure origin is visible
min_y = min(0, P1[1], P2[1], P3[1]) - 50


plt.xlim(min_x, max_x)
plt.ylim(min_y, max_y)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)

# Add labels and title (   )
plt.xlabel('  (V) / Real Part (V)')
plt.ylabel('  (V) / Imaginary Part (V)')
plt.title('      ( )\nPhasor Diagram for Ohmic Load (Nominal Loading)')
plt.legend(loc='upper left')
plt.gca().set_aspect('equal', adjustable='box') # Ensure aspect ratio is equal for correct angles

# Show plot ( )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#         3
# V_line = 150V ( ,  10   )
# Is = 4.7A ( ,  10   )
Rs = 2.44  #    ()
Xs = 101.3 #   ()

V_line = 150.0
Is = 4.7

#    V ( V_ph)
V_ph = V_line / np.sqrt(3) #  86.6V

#   
IsRs = Is * Rs     #  11.468V
IsXs = Is * Xs     #  476.11V

#   ,  V_ph   IsRs     (..  ).
#  IsXs   90 .
# EA = V_ph + IsRs + j*IsXs

#   
V_ph_vec = np.array([V_ph, 0])         #  V_ph ()
IsRs_vec = np.array([IsRs, 0])         #  IsRs
IsXs_vec = np.array([0, IsXs])         #  jIsXs

#   EA
# EA_real = V_ph + IsRs
# EA_imag = IsXs
EA_vec = np.array([V_ph + IsRs, IsXs])

#    
plt.figure(figsize=(8, 8))
ax = plt.gca() # Get current axes

#   
# 1. V_ph (    )
ax.arrow(0, 0, V_ph_vec[0], V_ph_vec[1], head_width=10, head_length=15, fc='blue', ec='blue', length_includes_head=True)
ax.text(V_ph_vec[0]/2, V_ph_vec[1] - 15, f'$V_{{ph}} = {V_ph:.1f}V$', color='blue', ha='center')

# 2. IsRs (     V_ph)
ax.arrow(V_ph_vec[0], V_ph_vec[1], IsRs_vec[0], IsRs_vec[1], head_width=10, head_length=15, fc='red', ec='red', length_includes_head=True)
ax.text(V_ph_vec[0] + IsRs_vec[0]/2, V_ph_vec[1] + 10, f'$I_sR_s = {IsRs:.1f}V$', color='red', ha='center')

# 3. jIsXs (     IsRs)
origin_IsXs_x = V_ph_vec[0] + IsRs_vec[0]
origin_IsXs_y = V_ph_vec[1] + IsRs_vec[1]
ax.arrow(origin_IsXs_x, origin_IsXs_y, IsXs_vec[0], IsXs_vec[1], head_width=10, head_length=15, fc='green', ec='green', length_includes_head=True)
ax.text(origin_IsXs_x + 20, origin_IsXs_y + IsXs_vec[1]/2, f'$jI_sX_s = j{IsXs:.1f}V$', color='green', va='center')

# 4. EA (         jIsXs)
ax.arrow(0, 0, EA_vec[0], EA_vec[1], head_width=10, head_length=15, fc='purple', ec='purple', length_includes_head=True, linestyle='--')
ax.text(EA_vec[0]*0.3, EA_vec[1]*0.7, f'$E_A = ({EA_vec[0]:.1f} + j{EA_vec[1]:.1f})V$', color='purple', ha='center')


#      
max_val_x = EA_vec[0] + 50
max_val_y = EA_vec[1] + 50
min_val_x = -20
min_val_y = -20

ax.set_xlim([min_val_x, max_val_x])
ax.set_ylim([min_val_y, max_val_y])
ax.set_aspect('equal', adjustable='box') #    
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle=':')

#   
plt.xlabel('  $Re$ (V)')
plt.ylabel('  $Im$ (V)')
plt.title('    $\cos\phi_R$ ( )')
plt.show()
```
