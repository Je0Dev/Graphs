```python
import matplotlib.pyplot as plt

# Δεδομένα για την αύξηση του ρεύματος διέγερσης
If_increase = [0, 0.02, 0.04, 0.06, 0.08, 0.10, 0.12, 0.14, 0.16, 0.18, 0.20]
U_increase = [40, 65, 100, 150, 210, 260, 310, 350, 380, 400, 420]
N_increase = 1495

# Δεδομένα για τη μείωση του ρεύματος διέγερσης
If_decrease = [0, 0.02, 0.04, 0.06, 0.08, 0.10, 0.12, 0.14, 0.16, 0.18, 0.20]
U_decrease = [70, 105, 165, 225, 280, 330, 350, 380, 395, 410, 420]
N_decrease = 1495

# Δημιουργία του γραφήματος
plt.figure(figsize=(10, 6))
plt.plot(If_increase, U_increase, marker='o', linestyle='--', color='blue', label='Αύξηση Ρεύματος Διέγερσης')
plt.plot(If_decrease, U_decrease, marker='o', linestyle='--', color='red', label='Μείωση Ρεύματος Διέγερσης')

# Προσθήκη τίτλου και ετικετών αξόνων
plt.title('Χαρακτηριστική Ανοικτού Κυκλώματος Εa=Uoc=f(If)')
plt.xlabel('Ρεύμα Διέγερσης If (A)')
plt.ylabel('Τάση Επαγωγής Ea = U (V)')

# Προσθήκη υπομνήματος
plt.legend()

# Προσθήκη πλέγματος για ευκολότερη ανάγνωση
plt.grid(True)

plt.scatter(If_increase, U_increase, color='green', zorder=2)
plt.scatter(If_decrease, U_decrease, color='purple', zorder=2)


# Ενημέρωση του υπομνήματος για να συμπεριλάβει και τα σημεία
plt.legend()

# Εμφάνιση του γραφήματος
plt.show()
```

```python
import matplotlib.pyplot as plt

# Δεδομένα Πίνακα 1: Ωμικό Φορτίο
U_ohm = [380, 370, 365, 355, 345, 330, 305, 280, 235, 150]
If_ohm = [0, 0.8, 1.0, 1.5, 2, 2.5, 3, 3.5, 4, 4.7]

# Δεδομένα Πίνακα 2: Επαγωγικό Φορτίο (συνφ=0,8)
U_ind = [380, 345, 330, 305, 285, 250, 220, 185]
If_ind = [0, 1.10, 1.60, 2.00, 2.30, 2.90, 3.20, 3.70]

# Δεδομένα Πίνακα 3: Χωρητικό Φορτίο (συνφ=0,8)
U_cap = [380, 403, 408, 406, 402, 400]
If_cap = [0, 1.6, 2.5, 3.2, 3.7, 4.1]

# Δημιουργία του γραφήματος
plt.figure(figsize=(12, 7))

# Σχεδίαση της χαρακτηριστικής για ωμικό φορτίο
plt.plot(If_ohm, U_ohm, marker='o', linestyle='--', color='blue', label='Ωμικό Φορτίο (cosφ_R=1)')
plt.scatter(If_ohm, U_ohm, color='purple',zorder=2)

# Σχεδίαση της χαρακτηριστικής για επαγωγικό φορτίο
plt.plot(If_ind, U_ind, marker='o', linestyle='--', color='red', label='Επαγωγικό Φορτίο (cosφ_L=0.8 Yστερεί)')
plt.scatter(If_ind, U_ind, color='black',zorder=2)

# Σχεδίαση της χαρακτηριστικής για χωρητικό φορτίο
plt.plot(If_cap, U_cap, marker='o', linestyle='--', color='green', label='Χωρητικό Φορτίο (cosφ_C=0.8 Προηγείται)')
plt.scatter(If_cap, U_cap, color='palevioletred',zorder=2)

# Προσθήκη τίτλου και ετικετών αξόνων
plt.title('Χαρακτηριστικές Φόρτισης Γεννήτριας Leroy Somer')
plt.xlabel('Ρεύμα Φορτίου IΦ [A]')
plt.ylabel('Τάση Εξόδου U [V]')

# Προσθήκη υπομνήματος
plt.legend()

# Προσθήκη πλέγματος για ευκολότερη ανάγνωση
plt.grid(True)

# Εμφάνιση του γραφήματος
plt.show()
```
