```python
import matplotlib.pyplot as plt
import numpy as np

#     V ()
field_current = np.linspace(0.1, 2.0, 100) #  
armature_current_nominal = 1.5 * (field_current - 1)**2 + 0.5 #     
armature_current_50_percent = 0.8 * (field_current - 1)**2 + 0.3 #    50% 

plt.figure(figsize=(8, 6))
plt.plot(field_current, armature_current_nominal, label='  (100%)')
plt.plot(field_current, armature_current_50_percent, label='50%  ')

#    (  )
min_current_idx_nominal = np.argmin(armature_current_nominal)
min_current_idx_50_percent = np.argmin(armature_current_50_percent)
plt.axvline(x=field_current[min_current_idx_nominal], color='gray', linestyle='--', label='Cos()=1 ()')
plt.axvline(x=field_current[min_current_idx_50_percent], color='grey', linestyle=':', label='Cos()=1 (50% )')

#  
plt.text(0.3, 1.8, ' ()', rotation=90, color='red')
plt.text(1.7, 1.8, ' ()', rotation=-90, color='blue')

plt.xlabel('  $I_f$ ()')
plt.ylabel('  $I_a$ ()')
plt.title('V-  ')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Synchronous motor characteristics
synchronous_speed_rpm = 1500  # Example synchronous speed (RPM)
synchronous_speed_rad_s = synchronous_speed_rpm * (2 * np.pi / 60) # rad/s

# Torque values (example values)
rated_torque = 500  # Example rated torque (Nm)
pull_out_torque = 1.8 * rated_torque # Example: Pull-out torque is typically 1.5 to 2.5 times rated torque
pull_in_torque = 0.8 * pull_out_torque # Example: Pull-in torque is less than pull-out torque, often around 0.8 of it

# Create plot
plt.figure(figsize=(8, 6))

# Plot the ideal torque-speed characteristic (vertical line at synchronous speed)
plt.plot([synchronous_speed_rpm, synchronous_speed_rpm], [0, pull_out_torque * 1.1], 'b-', linewidth=2, label='Ideal Torque-Speed Characteristic')

# Mark the characteristic torque values
plt.plot(synchronous_speed_rpm, pull_in_torque, 'go', markersize=8, label=f'Pull-in Torque: {pull_in_torque:.1f} Nm')
plt.plot(synchronous_speed_rpm, pull_out_torque, 'ro', markersize=8, label=f'Pull-out Torque: {pull_out_torque:.1f} Nm')

# Add labels and title
plt.xlabel(' (RPM)')
plt.ylabel(' (Nm)')
plt.title(' -  ')
plt.grid(True)
plt.legend()
plt.ylim(bottom=0)
plt.xlim(left=0)
plt.text(synchronous_speed_rpm + 50, pull_out_torque * 0.9, f' : {synchronous_speed_rpm} RPM',
         verticalalignment='center', horizontalalignment='left', color='blue')

plt.show()
```
