```python
import matplotlib.pyplot as plt
import numpy as np

#      V
If = np.linspace(0.5, 2.5, 200) #  

#  100% 
Ia_100 = np.sqrt(1.2**2 + (1.5/If - 1.5)**2)
min_Ia_100_idx = np.argmin(Ia_100)
If_min_100 = If[min_Ia_100_idx]

#  50% 
Ia_50 = np.sqrt(0.6**2 + (1.5/If - 1.5)**2)
min_Ia_50_idx = np.argmin(Ia_50)
If_min_50 = If[min_Ia_50_idx]

plt.figure(figsize=(10, 6))
plt.plot(If, Ia_100, label='100%  ')
plt.plot(If, Ia_50, label='50%  ')

#    cos=1
plt.plot(If_min_100, Ia_100[min_Ia_100_idx], 'ro')
plt.plot(If_min_50, Ia_50[min_Ia_50_idx], 'bo')
plt.text(If_min_100, Ia_100[min_Ia_100_idx] + 0.05, 'cos=1', horizontalalignment='center')
plt.text(If_min_50, Ia_50[min_Ia_50_idx] + 0.05, 'cos=1', horizontalalignment='center')

#  
plt.axvspan(0.5, If_min_50, color='orange', alpha=0.2, label=' ()')
plt.axvspan(If_min_100, 2.5, color='cyan', alpha=0.2, label=' ()')

plt.title(' V  ')
plt.xlabel('  If (A)')
plt.ylabel('  Ia (A)')
plt.grid(True)
plt.legend()
plt.ylim(0, 2.5)
plt.xlim(0.5, 2.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ns = 1000 # RPM
nominal_torque = 1050.4

plt.figure(figsize=(8, 6))
plt.plot([ns, ns], [0, 1.5], 'r-', linewidth=3, label=f'   {ns} RPM')
plt.plot(ns, 1.2, 'ro')
plt.text(ns - 150, 1.2, f'T_N  {nominal_torque:.2f} Nm', verticalalignment='center')

plt.title(' -  ')
plt.xlabel(' (RPM)')
plt.ylabel(' ()')
plt.xticks(np.arange(0, 1201, 200))
plt.yticks([])
plt.grid(True, axis='x')
plt.legend()
plt.xlim(0, 1200)
plt.ylim(0, 2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

s = np.linspace(0.001, 1, 200) # 
speed_rpm = 1500 * (1-s) #    1500 RPM
k = 2.5
X2 = 0.5

R2_values = {'R_ = R1 ()': 0.2, 'R_ = R2 ()': 0.5, 'R_ = R3 ()': 1.0}

plt.figure(figsize=(10, 6))
for label, R2 in R2_values.items():
    T = (k * s * R2) / (R2**2 + (s*X2)**2)
    plt.plot(speed_rpm, T, label=label)

plt.title(' -    ')
plt.xlabel(' (RPM)')
plt.ylabel(' ()')
plt.grid(True)
plt.legend()
plt.xlim(0, 1500)
plt.ylim(0, 3)
plt.show()
```
