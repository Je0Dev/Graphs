```python
import numpy as np
import matplotlib.pyplot as plt

# Τιμές εξαρτημάτων
L2 = 30e-3  # Henry
C2 = 0.02e-6 # Farad
R2 = 10e3    # Ohm

# Δεδομένα από τον πίνακα
f_kHz_data = np.array([1, 2, 3, 3.3, 3.6, 3.9, 4.2, 4.5, 4.8, 5, 6, 7, 8])
f_Hz_data = f_kHz_data * 1000
Up_V_data = np.array([0.10, 0.24, 0.55, 0.76, 1.03, 1.38, 1.38, 1.15, 0.88, 0.77, 0.43, 0.30, 0.24])

# Δημιουργία εύρους συχνοτήτων για τη σύνθετη αντίσταση
f_Hz_dense = np.linspace(1000, 10000, 500)
f_kHz_dense = f_Hz_dense / 1000  # Δημιουργία αντίστοιχου array σε kHz
omega_dense = 2 * np.pi * f_Hz_dense

# Υπολογισμός της σύνθετης αντίστασης του παράλληλου LC
Z_LC_mag = np.abs((omega_dense * L2) / (1 - omega_dense**2 * L2 * C2))

# Σχεδίαση
plt.figure(figsize=(12, 6))

# Σχεδίαση της σύνθετης αντίστασης
plt.plot(f_kHz_dense, Z_LC_mag, label='$|Z_{LC}|$ ($\Omega$)', color='blue')

# Σχεδίαση της τάσης U_Π (V_L = V_C)
plt.plot(f_kHz_data, Up_V_data, label='$V_L = V_C = U_\Pi$ (V)', color='red', linestyle='--')

plt.xlabel('Συχνότητα (kHz)')
plt.ylabel('Μέτρο Σύνθετης Αντίστασης ($\Omega$), Τάση (V)')
plt.title('Σύνθετη Αντίσταση Παράλληλου LC και Τάση Συντονισμού Συναρτήσει της Συχνότητας')
plt.grid(True)
plt.legend()
plt.show()
```
