```python
import numpy as np
import matplotlib.pyplot as plt

# Δεδομένα
f = 1  # Hz (συχνότητα)
T = 1 / f  # Περίοδος
Vpp = 4  # Peak-to-Peak τάση
Vdc = 1  # Μέση τιμή
Vrms=2.24

# Υπολογισμός πλάτους ημιτονοειδούς σήματος
A = Vpp / 2  # Peak Amplitude

# Χρονική κλίμακα
t = np.linspace(0, 2 * T, 1000)  # Δύο περίοδοι

# Ημιτονοειδές σήμα με μέση τιμή Vdc
v_t = A * np.cos(2 * np.pi * f * t) + Vdc

# Σχεδίαση γραφικής παράστασης χωρίς την υπολογισμένη τιμή του Vrms
plt.figure(figsize=(8, 4))
plt.plot(t, v_t, label=r'$Vt) = A\cos(2\pi ft) + V_{dc}$', color='b')
plt.axhline(Vdc, color='r', linestyle='--', label=r'$V_{dc}$')
plt.axhline(Vrms, color='g', linestyle='--', label=r'$V_{rms}$')
plt.xlabel('t(s)')
plt.ylabel('V(t) (V)')
plt.title('Συνημιτονοειδές σήμα')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def vt(t):
  """Defines the voltage function v(t) = 5cos(t) - 3."""
  return 5 * np.cos(t) - 3

def calculate_vdc(v, T):
  """Calculates the DC component (Vdc) of a periodic signal."""
  t = np.linspace(0, T, 1000)  # Use more points for better accuracy
  return np.mean(v(t))

def calculate_vrms(v, T):
  """Calculates the RMS value (Vrms) of a periodic signal."""
  t = np.linspace(0, T, 1000)
  return np.sqrt(np.mean(v(t)**2))

# Define the period (T) - since it's cos(t), the period is 2*pi
T = 2 * np.pi

# Calculate Vdc and Vrms
vdc = calculate_vdc(vt, T)
vrms = calculate_vrms(vt, T)

# Generate time values for plotting
t = np.linspace(0, 3 * T, 1000)  # Plot a few periods
y = vt(t)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t, y, label='v(t) = 5cos(t) - 3')

# Plot Vdc as a horizontal line
plt.axhline(y=vdc, color='r', linestyle='--', label=f'Vdc = {vdc:.2f} V')

# Plot Vrms as a horizontal line
plt.axhline(y=vrms, color='g', linestyle='--', label=f'Vrms = {vrms:.2f} V')

plt.xlabel('Time (t)')
plt.ylabel('Voltage (v)')
plt.title('Συνημιτονοειδές σήμα')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def vt(t, A, f, phi):
    """Defines the voltage function v(t) = A*cos(2*pi*f*t + phi)."""
    return A * np.cos(2 * np.pi * f * t + phi)

# Define parameters
A = 1  # Amplitude (μπορείτε να το αλλάξετε)
f = 1  # Frequency (μπορείτε να το αλλάξετε)
T = 1 / f  # Period

# Phase values from the table
phase_values = [-3 * np.pi / 4, -np.pi / 4, -np.pi / 6, 0, np.pi / 4, np.pi / 2, 3 * np.pi / 4]

# List of colors for each plot
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black']

# Create plots for each phase value
for i, phi in enumerate(phase_values):
    plt.figure(figsize=(10, 6))
    t = np.linspace(0, 3 * T, 1000)  # Plot a few periods
    y = vt(t, A, f, phi)
    plt.plot(t, y, color=colors[i], label=f'φ = {phi:.2f} rad')
    plt.xlabel('Time (t)')
    plt.ylabel('Voltage (v)')
    plt.title(f'Voltage Signal with φ = {phi:.2f} rad')
    plt.grid(True)
    plt.legend()
    plt.show()
```
